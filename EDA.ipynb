```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr

wr.filterwarnings('ignore')
```

```python
df = pd.read_excel("wine_data.xlsx")

df.head()
```

```python
print(df.shape)

print(df.info())

print(df.describe().T)

print(df.columns.tolist())

print(df.isnull().sum())

print(df.duplicated().sum())
```

```python
#BAR PLOT
quality_counts = df['quality'].value_counts()

plt.figure(figsize=(8, 6))

plt.bar(
    quality_counts.index,
    quality_counts,
    color='deeppink'
)

plt.title('Count Plot of Quality')
plt.xlabel('Quality')
plt.ylabel('Count')

plt.show()
```

```python
#HISTOGRAM
sns.set_style("darkgrid")

numerical_columns = df.select_dtypes(
    include=["int64", "float64"]
).columns

plt.figure(figsize=(12, 8))

for idx, feature in enumerate(numerical_columns, 1):

    plt.subplot(len(numerical_columns), 2, idx)

    sns.histplot(
        df[feature],
        kde=True
    )

    plt.title(f"{feature} Distribution")

plt.tight_layout()

plt.show()
```

```python
#SWARM PLOT
plt.figure(figsize=(8, 6))

sns.swarmplot(
    x="quality",
    y="alcohol",
    data=df,
    palette='viridis'
)

plt.title('Swarm Plot')
plt.xlabel('Quality')
plt.ylabel('Alcohol')

plt.show()
```

```python
#PAIR PLOT
g = sns.pairplot(df)

g.fig.suptitle('Pair Plot', y=1.02)

plt.show()
```

```python
#Violin Plot
df['quality'] = df['quality'].astype(str)

plt.figure(figsize=(8, 6))

sns.violinplot(
    x="quality",
    y="alcohol",
    data=df,
    palette='Pastel1'
)

plt.title('Violin Plot')
plt.xlabel('Quality')
plt.ylabel('Alcohol')

plt.show()
```

```python
#BOX PLOT
plt.figure(figsize=(8, 6))

sns.boxplot(
    x='quality',
    y='alcohol',
    data=df
)

plt.title('Box Plot')

plt.show()
```

```python
#Heatmap
plt.figure(figsize=(8, 6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```
